```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
CHUNK_SIZE = 1000000
chunk = pd.read_csv('instagram_posts.csv',sep='\t',chunksize=CHUNK_SIZE)
for data in chunk:
    data["date_data"] = (pd.to_datetime(data["cts"])).apply(lambda x: x.date())
    # print(type(data["date_data"][0]))
    two_rows = (data.assign(count=data.description.str.count('pixel'))).groupby('date_data')['count'].sum()
    plt.figure(figsize=(20,20))
    two_rows[two_rows!=0].plot.line()
    # two_rows.plot.line()
    break
```
